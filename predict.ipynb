```python
import pandas as pd
```

```python
laliga_matches = pd.read_csv("matches_laliga.csv") 
```

```python
laliga_matches.dtypes
```

```python
laliga_matches['date'] = pd.to_datetime(laliga_matches['date'])
laliga_matches.dtypes
```

```python
# Home and Away advantage
laliga_matches['venue_code'] = laliga_matches['venue'].astype('category').cat.codes
# code for each opponent
laliga_matches['opponent_code'] = laliga_matches['opponent'].astype('category').cat.codes
# The hour and day of the match might affect results
laliga_matches['hour'] = laliga_matches['time'].str.split(':').str[0].astype(int)
laliga_matches['day_code'] = laliga_matches['date'].dt.dayofweek
# Some studies say referee bias might exist
laliga_matches['referee_code'] = laliga_matches['referee'].astype('category').cat.codes
# Captains might have an influence
laliga_matches['captain_code'] = laliga_matches['captain'].astype('category').cat.codes
# Match week can have an effect
laliga_matches['round_code'] = laliga_matches['round'].astype('category').cat.codes
# Formation definitely plays a big role
laliga_matches['form_code'] = laliga_matches['formation'].astype('category').cat.codes
laliga_matches['oppform_code'] = laliga_matches['opp formation'].astype('category').cat.codes
# Target variable: Win = 1, Lose or Draw = 0
laliga_matches['target'] = (laliga_matches['result'] == 'W').astype('int')
```

```python
laliga_matches
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train_data = laliga_matches[laliga_matches['date'] < '2025-01-01']
test_data = laliga_matches[laliga_matches['date'] >= '2025-01-01']
predictors = ['venue_code', 'opponent_code', 'hour', 'day_code', 'referee_code', 'captain_code','round_code', 'form_code', 'oppform_code']
rf.fit(train_data[predictors], train_data['target'])
predictions = rf.predict(test_data[predictors])
```

```python
acc = accuracy_score(test_data['target'], predictions)
print(f"Accuracy: {acc}")

combined = pd.DataFrame({
    "actual": test_data["target"], 
    "prediction": predictions
})

confusion_matrix = pd.crosstab(index=combined ["actual"], columns=combined ["prediction"])

precision = precision_score(test_data['target'], predictions)
print(f"Precision: {precision}")
print(confusion_matrix)
```

```python
# We want to use the stats of the previous 3 weeks to predict the next match
def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')
    rolling_states = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_states
    group = group.dropna(subset=new_cols)
    return group
```

```python
# Calculating the rolling averages for these status
cols = ['gf', 'ga', 'sh', 'sot', 'dist','poss']
new_cols = [f'{c}_rolling' for c in cols]
print(new_cols)

matches_rolling = laliga_matches.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling = matches_rolling.droplevel('team')
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
predictors_rolling = predictors + new_cols
training_set = matches_rolling[matches_rolling['date'] < '2025-01-01']
test_set = matches_rolling[matches_rolling['date'] > '2025-01-01']
rf.fit(training_set[predictors_rolling], training_set['target'])
predictions = rf.predict(test_set[predictors_rolling])
combined = pd.DataFrame(dict(actual=test_set['target'], predicted=predictions), index = test_set.index)
precision = precision_score(test_set['target'], predictions)
accuracy = accuracy_score(test_set['target'], predictions)
```

```python
print(f"Accuracy: {round(accuracy, 3)}\nPrecision: {round(precision, 3)}")
```
